## Step 15 복잡한 계산 그래프(이론 편)

#### 역전파의 올바른 순서
지금까지 만든 DeZero는 한줄로 늘어선 계산 그래프만 다룸
복잡하게 연결된 그래프의 역전파를 구하려면 순서가 중요
전 단계의 미분값이 모두 전파된 '후'에 다시 미분값을 전파해야 한다.

#### 현재의 DeZero
지금 만든 DeZero는 마지막 생성자를 리스트에 추가하고 꺼내서 전파하기 때문에 문제가 생긴다

#### 함수 우선순위
이를 해결하려면 함수 우선순위를 정해주어야 한다.

간단한 방법으로 
순전파를 계산할때 함수가 변수를 만들어내는 과정을 이미 알고있다.
즉 어떤 함수가 어떤 변수를 만들어내는가(부모-자식 관계)를 알고있다.
이 관계를 이용하여 변수의 세대 를 기록하면 우선순위로 쓸 수 있다.
